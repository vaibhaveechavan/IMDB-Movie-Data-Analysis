### 20. Classifying movies based on ratings into excellent, good and average

```python
def rating(rating):
    if rating  >= 7.0:
        return "Excellent"
    elif rating >= 6.0:
        return "Good"
    else:
        return "Average"
```

```python
data['Rating_Category'] = data['Rating'].apply(rating)
```

### 22. Finding unique values from Genre

```python
list1 = []
for value in data['Genre']:
    list1.append(value.split(','))
list1
```

```python
one_dl = []
for item in list1:
    for item1 in item:
        one_dl.append(item1)
one_dl
```

```python
uni_list = []
for item in one_dl:
    if item not in uni_list:
        uni_list.append(item)
uni_list
```

```python
len(uni_list)
```
